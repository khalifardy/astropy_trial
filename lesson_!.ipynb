```python
from astropy.io import fits
file = fits.open('502nmos.fits') #read file fits
print(file)
```

```python
image_data = file[0].data
print(image_data)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
plt.figure()
plt.imshow(image_data, origin='lower',norm = LogNorm(), cmap='Greys')
plt.colorbar()
plt.show()
```

```python
section1 = image_data[1000:1250, 1300:1350]
plt.figure()
plt.imshow(section1, origin='lower',norm = LogNorm(), cmap='Greys')
plt.colorbar()
plt.show()
```
